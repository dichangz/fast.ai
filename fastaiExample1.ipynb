# This file is for training and testing model
The fastai courses is using fastai V0.7 and this script is using fastai V1.0

```python
from fastai import *
from fastai.vision import *
```

```python
torch.cuda.is_available() and torch.backends.cudnn.enabled
```

```python
os.getcwd()
```

```python
PATH = os.path.join(os.getcwd(),'Pictures','fastaidata')
```

```python
data = ImageDataBunch.from_folder(PATH, ds_tfms = get_transforms(), size =224)
learn = create_cnn(data, models.resnet34, metrics=accuracy)
learn.fit(1)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

Then we can see the most incorrect items. The first one is what I should delete beforehand.

```python
interp.plot_top_losses(4, figsize=(20,20))
```

The confusion matrix

```python
interp.plot_confusion_matrix()
```

Then we can see the property of each item.

```python
img = learn.data.train_ds[0][0]
img.show(figsize=(2,2))
learn.predict(img)
```
